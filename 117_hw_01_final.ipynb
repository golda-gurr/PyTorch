#### Вариант первый (один шаг оптимизации)

```python
# importing libraries
from torch import nn 
from torch import optim
```

```python
x_data = Variable(torch.tensor([[2.0]]))
y_data = Variable(torch.tensor([[12.0]]))
w1 = Variable(torch.randn(1), requires_grad = True) #  creating random tensors for weights
```

```python
def forward(x):
  y = x**w1 - 2 * x**2 + 5
  return y
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD([w1], lr=0.001)
```

```python
optimizer.zero_grad()
y_pred = forward(x_data)
loss = criterion(y_pred, y_data)
print('loss: ', loss.item())
print ('dY/dw1: ', w1.grad) 
```

```python
loss.backward()
print ('dY/dw1: ', w1.grad)
print ('w1: ', w1)
```

```python
optimizer.step()
print ('w1: ', w1)
print ('y_pred: ', x_data**w1 - 2 * x_data**2 + 5)
```

#### Вариант второй

```python
# importing libraries
from torch import nn 
from torch import optim
```

```python
x_data = Variable(torch.tensor([[2.0]]))
y_data = Variable(torch.tensor([[12.0]]))
w1 = Variable(torch.randn(1), requires_grad = True) #  creating random tensors for weights
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD([w1], lr=0.001)
```

```python
# Training loop
for epoch in range(500):
    # 1) Forward pass: Compute predicted y by passing x to the model
    y_pred = x_data**w1 - 2 * x_data**2 + 5

    # 2) Compute and print loss
    loss = criterion(y_pred, y_data)
    if epoch % 10 == 0:
      print(f'Epoch: {epoch} | Loss: {loss.item()} | y_pred: {y_pred} ')

    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

#### Вариант третий (оптимизация с nn.Module)

```python
# importing libraries
from torch import nn 
from torch import optim
```

```python
x_data = Variable(torch.tensor([[2.0]]))
y_data = Variable(torch.tensor([[12.0]]))
```

```python
class FuncModel(torch.nn.Module):
    def __init__(self):
        """
        In the constructor we instantiate four parameters and assign them as
        member parameters.
        """
        super().__init__()
        self.w1 = torch.nn.Parameter(torch.randn(()))

    def forward(self, x):
        """
        In the forward function we accept a Tensor of input data and we must return
        a Tensor of output data. We can use Modules defined in the constructor as
        well as arbitrary operators on Tensors.
        """
        y_pred = x.pow(self.w1) - 2 * x.pow(2) + 5
        return y_pred

    def string(self):
        """
        Just like any class in Python, you can also define custom method on PyTorch modules
        """
        return f'y = x**{self.w1.item()} - 2 * x**2 + 5 '
```

```python
model = FuncModel()
```

```python
criterion = nn.MSELoss(reduction='sum')
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)
```

```python
# Training loop
for epoch in range(500):
    # 1) Forward pass: Compute predicted y by passing x to the model
    y_pred = model(x_data)

    # 2) Compute and print loss
    loss = criterion(y_pred, y_data)
    if epoch % 10 == 0:
      print(f'Epoch: {epoch} | Loss: {loss.item()} | y_pred: {y_pred}')

    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
print(f'Result: {model.string()}')
```

```python
# After training
x_var = Variable(torch.tensor([[4.0]]))
y_pred = model(x_var)
print("Prediction (after training)",  model(x_var).data[0][0].item())
```
